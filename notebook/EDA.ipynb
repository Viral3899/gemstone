```python
import matplotlib.pyplot as plt
import seaborn as sns
import random


numerical_colums = [col for col in df.columns if df[col].dtype !='O']
categorical_colums = [col for col in df.columns if df[col].dtype =='O']
```

```python
numerical_colums
```

```python
colors=plt.colormaps()
```

```python

for i in numerical_colums:
    plt.figure(figsize=(8,8))
    sns.histplot(data=df,x=i,palette=random.choice(colors),kde=True)
    plt.show()
                 
```

```python
categorical_colums
```

```python

for i in categorical_colums:
    plt.figure(figsize=(8,8))
    sns.countplot(data=df,x=i,palette=random.choice(colors))
    plt.show()
                 
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True),cmap=random.choice(colors),square=True,annot=True)
```
